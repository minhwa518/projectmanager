```python
menu = '''[★나도 단어왕★]
--------------------
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기'''

word_king = {}

while True:
    print('-'*23)
    print(menu)
    choice = input('-'*20 + ' 선택: ')

    if choice == '1':
        word = input('새 단어: ')
        if word in word_king:
            print('등록된 단어입니다.')
        else:
            meaning = input(f'{word}의 뜻: ')
            word_king[word] = meaning
            print("등록이 완료되었습니다.")
            print(word_king)

    if choice == '2':
        word = input('검색할 단어: ')
        found = False

        for key, value in word_king.items():
            if word == key or word == value:
                print(f'{{{key}:{value}}}')
                found = True
                break
        if not found:
            print('미등록 단어입니다.')

    if choice == '3':
        word = input('삭제할 단어를 입력하세요: ')
        if word in word_king:
            print(f'{{{word}:{word_king[word]}}} 삭제가 완료되었습니다.')
            word_king.pop(word)
            print(word_king)
        else:
            print('미등록 단어입니다.')


    if choice == '4':
        print(word_king)

    if choice == '0':
        break

    if choice not in {'1','2','3','4','0'}:
        print('메뉴에 있는 항목만 입력하세요')

print('나도 단어왕 프로그램을 종료합니다.')



```
